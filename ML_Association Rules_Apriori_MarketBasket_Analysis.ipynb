Apriori algorithm is a classical algorithm in data mining

This process/algorithm uses in mall such as D-Mart, Reliance art, Wall Mart etc. to increase marketing and sale purpose.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Data PreProcessing
market=pd.read_csv('Market_Basket_Optimisation.csv',header=None)
market.head()
```

```python
market.shape
```

```python
market.info()
```

```python
market.isnull().sum()
```

```python
# Convert data in list format Method:1
market.values.tolist()
```

```python
# Method: 2

transactions=[]
for i in range(0,7500):
    transactions.append([str(market.values[i,j]) for j in range(0,20)])
transactions
```

```python
# Install Apriori as apyori

#pip install apyori
```

```python
# Training Apriori on the dataset

from apyori import apriori
rules=apriori(transactions, min_support=0.003,min_confidence=0.2, min_lift=3,min_length=2)
```

```python
# Now we will see how our 
# Visualising the results

results=list(rules)
results
```

```python
results=pd.DataFrame(results)
results.head()
```

```python
X=[]
Y=[]
Z=[]
W=[]

mylist=results['ordered_statistics'].tolist()
for ele in mylist:
    X.append(ele[0][0])
    Y.append(ele[0][1])
    Z.append(ele[0][2])
    W.append(ele[0][3])   
```

```python
results['Base_item']=X
results['Item_add']=Y
results['Confidence']=Z
results['Lift']=W
```

```python
results.head()
```

```python
results.drop(['ordered_statistics'],axis=1,inplace=True)
results.head()
```
